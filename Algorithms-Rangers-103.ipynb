## In-Place Algorithms

#### Syntax

```python
# var[i],var[i + 1] = var[i + 1], var[i]
# Sometimes known as a swap algorithm
# in place --> we are altering the original list

def swap(alist,x,y,z):
    alist[x],alist[y],alist[z] = alist[z],alist[y],alist[x]
    return alist

my_list = [20,4,10]
print(f"Before swap: {my_list}")
swap(my_list,2,0,1)
print(f"After swap: {my_list}")
```

#### Out of Place Algorithm

```python
# out of place --> we are creating a new list or a list copy
# not swapping, we are reversing the list completely
# but it also copies to another place in memory

my_list = [20, 4, 10]

my_list_copy = my_list[::-1]
print(my_list_copy)

array = ['a', 'b', 'c', 'd']
new_array = ['a'] *len(array)
print(new_array)
print("Before", array)
length = len(array) - 1 # = 3

for i in range(length): # 0, 1, 2
    new_array[i] = array[length - i]
    print(new_array[i])
    
array = new_array
print("After", array)

```

## Two Pointers

#### Syntax

```python
# alist[left], alist[right] = alist[right], alist[left]
# use two pointers to swap, you can a while loop in most cases
# --> in-place

my_list2 = [1, 2, 3, 12, 9, 8, 4, 11, 22]

def twoPointers(alist):
    # creating pointers for the list
    left = 0 # left is index 0, beginning at the start of the list or left side
    right = len(alist) - 1 # right is an index, starting at the end of the list or right side
    while left <= right:
        alist[left], alist[right] = alist[right], alist[left]
        left += 1
        right -= 1
    return alist

twoPointers(my_list2)
```

# Sorting Algorithms

```python
# constant time O(1)
array = [1, 2, 3, 4]
print(array[0])


# linear time O(n)
if 2 in array:
    print("it's here")

# linear time O(n)
for num in array:
    print(num)

# O(n log n) time
array2 = [1, 5, 6, 8, 4]
sorted_array = sorted(array2)
array2.sort()

# quadratic time O(n^2)
array = [1, 2, 3, 4]
array2 = [1, 5, 6, 8, 4, 6, 7, 4, 5, 2, 4]
for x in array: # looping through one list
    if x in array2: # searching in another
        print(x)

# quadratic time O(n^2)   
for x in array2:
    for y in array2:
        print(x + y)

# basically any nested linear operation is going to increase time complexity exponentially

# cubic --> exponential
for x in array2:
    for y in array2:
        for z in array2:
            print(x, y, z)
```

#### Bubble Sort

Worst Case: O(n^2) Time - O(1) Space

```python
# time complexity
# how many action have to happen per things taken as inputs

# .count() linear
# .append() constant
# .sort()/ sorted() O(n log n)
# .pop() constant if no index used
# .pop(4) linear
```

```python
# best case O(n) linear
# worst case O(n^2) quadratic

# after each pass one number will reach its final position

my_list = [22, 55, 88, 44, 1, 100, 34, 66]

def swap(i, j, array):
    array[i], array[j] = array[j], array[i]

def bubbleSort(array):
    isSorted = False # <-- setting a flag
    while not isSorted:
        isSorted = True
        for index in range(len(array) - 1):
            if array[index] > array[index + 1]:
                swap(index, index + 1, array)
                isSorted = False
    return array

bubbleSort(my_list)           
```

##### Insertion Sort

Worst Case: O(n^2) time - O(1)space

```python
my_list = [22, 55, 88, 44, 1, 100, 34, 66]

def swap(i, j, array):
    array[i], array[j] = array[j], array[i]

def insertionSort(array):
    for i in range(1, len(array)): # <-- considers 0 to be sorted
        # creating a check so we don't go past the 0 index
        j = i
        while j > 0 and array[j] < array[j - 1]:
            swap(j, j - 1, array)
            j -= 1
    return array
            
insertionSort(my_list)
```

## Merge Sort

```python
# O(n log n) time
# O(n) space
```

#### How it Works

```python
# step 1: split everything into its own group
# step 2: from the left to right merge two groups together
# step 3: while merging, place each item in the correct position within the merged group
# step 4: continue steps 3 & 4 until only one group is left

from random import randint
# generate a random list of 5 numbers between 0 to 20
nums = [randint(0, 20) for i in range(5)]

print(nums)

# merge sort
def mergeSort(alist):
    print("Splitting....", alist)
    
    # step 1: divide the array into equal parts (as much as possible)
    if len(alist) > 1:
        mid = len(alist) // 2
        lefthalf = alist[:mid]
        righthalf = alist[mid:]
        
        # recursively call mergeSort to perform splits until len(alist) == 1
        # merge once the splits are done
        mergeSort(lefthalf)
        mergeSort(righthalf)
        
        # index pointers for our list
        l = 0 # pointer for lefthalf
        r = 0 # pointer for righthalf
        m = 0 # pointer for main array
        
        # step 2: compare lefthalf with righthalf
        while l < len(lefthalf) and r < len(righthalf):
            if lefthalf[l] < righthalf[r]:
                alist[m] = lefthalf[l]
                l += 1
            else:
                alist[m] = righthalf[r]
                r += 1
            m += 1

        # step 3: while merging place items in correct position
        while l < len(lefthalf):
            alist[m] = lefthalf[l]
            l += 1
            m += 1
        
        while r < len(righthalf):
            alist[m] = righthalf[r]
            r += 1
            m += 1
    
    print("Merging: ", alist)
    return alist

mergeSort(nums)
```

# Binary Search

The Binary Search algorithm works by finding the number in the middle of a given array and comparing it to the target. Given that the array is sorted

* The worst case run time for this algorithm is `O(log(n))`

```python
# less == left
# greater == right
# list of numbers must be sorted

def binHelper(array, target, left, right):
    while left <= right:
        middle = (left + right) // 2
        potential_match = array[middle]
        if target == potential_match:
            return f"The index is....{middle}"
        elif target < potential_match:
            right = middle - 1
        else:
            left = middle + 1
    return -1

def binarySearch(array, target):
    return binHelper(array, target, 0, len(array)-1)

binarySearch([22, 44, 55, 66, 88, 100], 44)
```

### Exercise #1 <br>
<p>Reverse the list below in-place using an in-place algorithm.<br>For extra credit: Reverse the strings at the same time.</p>

```python
#1 extra extra credit will be if you can figure out the function for a list of ANY size. Not just this guy

# Version 1

alist_dict = {
    "alist1": ['this' , 'is', 'a', 'sentence', '.'],
    "alist2": ['this' , 'is', 'a', 'sentence', 'of', 'a', 'different', 'size', '.'],
    "alist3": ['this' , 'is', 'a', 'different', 'sentence', 'of', 'yet', 'another', 'size', '.'] 
}

def listReverse(alist):
    """
    Function that takes a list of any size and reverses it in place.
    Returns a list.
    """
    print(f"--> Memory location before: {id(alist)}")
    l = 0
    r = len(alist) - 1
    while l <= r:
        alist[l], alist[r] = alist[r], alist[l]
        l += 1
        r -= 1
    print(f"--> Memory location after: {id(alist)}")
    return alist


for value in alist_dict.values():
    print(f"List before: {value}")
    print(f"List after: {listReverse(value)}")
    print("\n")
```

```python
# For extra credit: Reverse the strings at the same time.
#1 extra extra credit will be if you can figure out the function for a list of ANY size. Not just this guy

# Version 2

alist_dict = {
    "alist1": ['this' , 'is', 'a', 'sentence', '.'],
    "alist2": ['this' , 'is', 'a', 'sentence', 'of', 'a', 'different', 'size', '.'],
    "alist3": ['this' , 'is', 'a', 'different', 'sentence', 'of', 'yet', 'another', 'size', '.'] 
}

def listReverse(alist):
    """
    Function that takes a list of any size and reverses it in place all at once.
    Returns a list.
    """
    print(f"--> Memory location before: {id(alist)}")
    length = len(alist)
    l = ""
    r = ""
    for i in range(length):
        l += f"alist[{i}],"
        r += f"alist[{(length-1)-i}],"
    # Now reverse all at once
    l_r = f"{l} = {r}"
    exec(l_r)
        
    print(f"--> Memory location after: {id(alist)}")
    return alist


for value in alist_dict.values():
    print(f"List before: {value}")
    print(f"List after: {listReverse(value)}")
    print("\n")
```

### Exercise #2 <br>
<p>Create a function that counts how many distinct words are in the string below, then outputs a dictionary with the words as the key and the value as the amount of times that word appears in the string.<br>Should output:<br>{'a': 5,<br>
 'abstract': 1,<br>
 'an': 3,<br>
 'array': 2, ... etc...</p>

```python
a_text = 'In computing, a hash table hash map is a data structure which implements an associative \
array abstract data type, a structure that can map keys to values. A hash table uses a hash function \
to compute an index into an array of buckets or slots from which the desired value can be found'

def wordCountDict(text):
    """
    Function that takes a string and counts how many distinct words are in the string.
    Returns a dictionary with the words as the key and the value as the amount of times 
    that the word appears in the string.
    """
    text = text.replace(",", "")
    text_list = text.split(" ")
    text_list = list(map(lambda x: x.lower().strip(), text_list))    
    
    word_dict = {}
    
    for word in sorted(set(text_list)):
        word_dict[word] = 0
    
    for word in text_list:
        word_dict[word] += 1
    
    return word_dict


wordCountDict(a_text)
```

```python
# this solution is in linear time O(n) since the algorithm only has to scan the list once for the search item

from random import randint
nums_list = [randint(0, 30) for i in range(11)]
print(f"Random list of 11 numbers: {nums_list}")
num = randint(0, 30)
print(f"Look for {num} in the random list of 11 numbers.  Return -1 if not found.")

def linearSearch(array, item):
    """
    Function that takes a list and an item to be found.
    Returns the item if found or -1 if not found.
    """
    if item in array:
        return item
    else:
        return -1


linearSearch(nums_list, num)
```
